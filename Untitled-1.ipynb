```python
import pandas as pd

movies = pd.read_csv('imdb_top_1000.csv')
```

```python
movies.head(1)
```

```python
movies.drop(['Poster_Link', 'Overview', 'Meta_score', 'No_of_Votes'], axis=1, inplace=True)
```

```python
movies.head(10)
```

```python
movies.dropna(inplace=True)
```

```python
movies.head(25)
```

```python
movies['Genre']=movies['Genre'].str.split(',').str[0]
```

```python
movies['Runtime']=movies['Runtime'].str.split(' ').str[0]
```

```python
movies.head(10)
```

```python
movies_clean = movies.drop(['Genre', 'Director', 'Star1', 'Star2', 'Star3', 'Star4'], axis=1)
```

```python
movies_clean.head(10)
```

```python
movies_clean.to_csv('movies.csv', index=False)
```

```python
actors = movies.drop(['Released_Year', 'Certificate', 'Runtime', 'IMDB_Rating', 'Gross', 'Director', 'Genre'], axis=1)
```

```python
actors.head(10)
```

```python
actors.to_csv('actors.csv', index=False)
```
